## A. Assess Baseline Model

```python
# DO NOT MODIFY THE CODE IN THIS CELL
# Load data
try:
  X_train = pd.read_csv(at.folder_path / 'X_train.csv')
  y_train = pd.read_csv(at.folder_path / 'y_train.csv')

  X_val = pd.read_csv(at.folder_path / 'X_val.csv')
  y_val = pd.read_csv(at.folder_path / 'y_val.csv')

  X_test = pd.read_csv(at.folder_path / 'X_test.csv')
  y_test = pd.read_csv(at.folder_path / 'y_test.csv')
except Exception as e:
  print(e)
```

### A.1 Generate Predictions with Baseline Model

```python
from sklearn.dummy import DummyRegressor

base_reg = DummyRegressor()
base_reg.fit(X_train, y_train)
```

### A.2 Selection of Performance Metrics

> Provide some explanations on why you believe the performance metrics you chose is appropriate


```python
performance_metrics_explanations = """
For predicting the target variable, net_premium_amount, I chose Mean Absolute Error (MAE) as the primary performance metric.

MAE measures the average magnitude of prediction errors, without considering their direction. It is expressed in the same units as the target (currency), making it directly interpretable and easy to communicate how far off, on average, predictions are from the actual values.

I opted for MAE over metrics like RMSE because the target distribution is relatively balanced(as I analyzed the target in EDA), with mean and median close together and only mild skewness. There are no extreme outliers that would necessitate penalizing larger errors more heavily, so MAE provides a robust and understandable measure of model performance."""
```

```python
# DO NOT MODIFY THE CODE IN THIS CELL
print_tile(size="h3", key='performance_metrics_explanations', value=performance_metrics_explanations)
```

### A.3 Baseline Model Performance

> Provide some explanations on model performance


```python

# Evaluate baseline model (Dummy Regressor) performance
y_train_preds = base_reg.predict(X_train)
y_val_preds = base_reg.predict(X_val)
y_test_preds = base_reg.predict(X_test)

# Collect results in a dictionary
results = {
    "Dataset": ["Training", "Validation", "Test"],
    "MAE": [
        mae(y_train, y_train_preds),
        mae(y_val, y_val_preds),
        mae(y_test, y_test_preds),
    ]
}

# Convert to DataFrame for a clean table
results_df = pd.DataFrame(results)
print(results_df)
```

```python
# <Student to fill this section and then remove this comment>
baseline_performance_explanations = """
The baseline model (dummy regressor predicting the mean net_premium_amount) gives us a reference point for performance. On the training set, the MAE is 38.09, which is relatively low because the model essentially predicts the central value of the target distribution. On the test set, the MAE rises to 232.29, showing that the model struggles to capture the individual variability in net_premium_amount.

This large difference between train and test errors is expected for a naive model that does not use any features. It highlights that the target has significant variability, which cannot be captured by a simple mean prediction. This baseline helps us understand the minimum level of error and sets a benchmark to evaluate more complex predictive models."""
```
